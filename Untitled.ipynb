# Customer Churn Prediction

### Objective:
To predict whether a customer will discontinue (or “churn”) a subscription-based service. The solution will not only classify the likelihood of churn but also offer insights into what factors are the most significant drivers behind customer attrition.

### Business Context:
Understanding churn helps businesses proactively address issues, optimize marketing and retention strategies, and ultimately reduce revenue loss due to customer turnover.



## 1. Load and Explore Data

```python
import pandas as pd

# Load data
df = pd.read_csv('Churn_Modelling.csv') 
```

```python
print(df.info())
```

```python
print(df.describe())
```

## 2. Clean & Preprocess Data

```python
# Drop uninformative columns
df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)
```

```python
# Encode categorical columns
from sklearn.preprocessing import LabelEncoder

df['Gender'] = LabelEncoder().fit_transform(df['Gender'])  # Male=1, Female=0
df['Geography'] = LabelEncoder().fit_transform(df['Geography'])  # France=0, Germany=1, Spain=2
```

```python
# Fill missing values
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
# Features & target
X = df.drop('Exited', axis=1)
y = df['Exited']
```

```python
# Scale features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## 3. Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Churn distribution
sns.countplot(x='Exited', data=df)
plt.title("Churn Distribution")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation")
plt.show()
```

```python
# Boxplots for numerical features
for col in ['CreditScore', 'Age', 'Balance', 'EstimatedSalary']:
    sns.boxplot(x='Exited', y=col, data=df)
    plt.title(f'{col} vs Churn')
    plt.show()
```

```python
# Countplot for categorical features
sns.countplot(x='Geography', hue='Exited', data=df)
plt.title('Geography vs Churn')
plt.show()

sns.countplot(x='Gender', hue='Exited', data=df)
plt.title('Gender vs Churn')
plt.show()
```

## 4. Train/Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, stratify=y, random_state=42)
```

## 5. Model Training and Evaluation

We'll train and compare the following models:

   1. Logistic Regression
    
   2. Random Forest
    
   3. XGBoost
    
   4. Support Vector Machine (SVM)
    
   5. K-Nearest Neighbors (KNN)
    
   6. Neural Network (MLPClassifier)

### Define Utility Functions

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve

def evaluate_model(model, name):
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None
    
    print(f"\n{name} Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    
    if y_proba is not None:
        auc = roc_auc_score(y_test, y_proba)
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        plt.plot(fpr, tpr, label=f'{name} (AUC = {auc:.2f})')
```

### Train Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
import xgboost as xgb

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "XGBoost": xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    "SVM": SVC(probability=True),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Neural Network": MLPClassifier(max_iter=500)
}

plt.figure(figsize=(10, 6))
for name, model in models.items():
    model.fit(X_train, y_train)
    evaluate_model(model, name)
plt.plot([0, 1], [0, 1], 'k--')
plt.title('ROC Curves')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

## 6. Feature Importance (Tree-based Models)

```python
# Random Forest
importances = models["Random Forest"].feature_importances_
feature_names = X.columns

plt.figure(figsize=(10, 6))
sns.barplot(x=importances, y=feature_names)
plt.title("Random Forest Feature Importance")
plt.show()
```

```python
# XGBoost
xgb.plot_importance(models["XGBoost"], height=0.6)
plt.title("XGBoost Feature Importance")
plt.show()
```

## 7. Model Comparison Summary Table

```python
results = {}
for name, model in models.items():
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else [0] * len(y_pred)
    
    results[name] = {
        'Accuracy': accuracy_score(y_test, y_pred),
        'ROC AUC': roc_auc_score(y_test, y_proba) if hasattr(model, "predict_proba") else None
    }

results_df = pd.DataFrame(results).T.sort_values(by="Accuracy", ascending=False)
print("\nModel Comparison Summary:\n")
print(results_df)
```
